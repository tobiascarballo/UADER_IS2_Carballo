```python
import numpy as np
import matplotlib.pyplot as plt

#primero, calculo del esfuerzo E segun el tamao S
S = np.linspace(0, 10000, 500) # 500 valores entre 0 y 10.000
E = 8 * (S ** 0.95)

#segndo, calculo del tiempo calendario td segun el esfuerzo
E_range = np.linspace(1, 500, 500) #500 valores entre 1 y 500
td = 2.4 * (E_range ** 0.33)

#tercero, grafico de E(esfeuerzo) vs S(tamao)
plt.figure()
plt.plot(S, E)
plt.xlabel("Tamao del proyecto S")
plt.xlabel("Esfuerzo E")
plt.title("Relacion entre tamao y esfuerzo")
plt.grid(True)
plt.show()

#cuarto, grafico de TD(tiempo calendario) vs E(esfuerzo)
plt.figure()
plt.plot(E_range, td)
plt.xlabel("Esfuerzo E")
plt.xlabel("Tiempo calendario td")
plt.title("Relacion entre esfuerzo y tiempo calendario")
plt.grid(True)
plt.show()
```
